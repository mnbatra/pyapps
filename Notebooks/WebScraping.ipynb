```python
import requests
from bs4 import BeautifulSoup
import pandas
l=[]
base_url="https://pythonhow.com/real-estate/rock-springs-wy/LCWYROCKSPRINGS/#t=0&s="
for i in range(0,30,10):
    r=requests.get(base_url+str(i))
    c=r.content

    soup=BeautifulSoup(c,"html.parser")
    #print(soup.prettify())
    all=soup.find_all("div",{"class":"propertyRow"})
    for items in all:
        d={}
        if "0 Gateway" in items.find_all("span",{"class":"propAddressCollapse"})[0].text:
            continue
        else:
            d["Address"]=items.find_all("span",{"class":"propAddressCollapse"})[0].text
        try:
            d["Locality"]=items.find_all("span",{"class":"propAddressCollapse"})[1].text
        except:
            d["Locality"]=None
        d["Price"]=items.find("h4",{"class":"propPrice"}).text.replace("\n","").replace(" ","")

        try:
            d["Beds"]=items.find("span",{"class":"infoBed"}).find("b").text
        except:
            d["Beds"]=None
        try:
            d["Area"]=items.find("span",{"class":"infoSqFt"}).find("b").text
        except:
            d["Area"]=None
        try:
            d["Full Baths"]=items.find("span",{"class":"infoValueFullBath"}).find("b").text
        except:
            d["Full Baths"]=None
        try:
            d["Half Baths"]=items.find("span",{"class":"infoValueHalfBath"}).find("b").text
        except:
            d["Half Baths"]=None

        for column_group in items.find_all("div",{"class":"columnGroup"}):
            for feature_group,feature_name in zip(column_group.find_all("span",{"class":"featureGroup"}),column_group.find_all("span",{"class":"featureName"})):
                #print(feature_group.text,feature_name.text)
                if "Lot Size" in feature_group.text:
                    d["Lot Size"]=feature_name.text
        l.append(d)
df=pandas.DataFrame(l)
df.to_csv("PropertyScrap.csv")
df
```
